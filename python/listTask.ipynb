```python
#Check whether {"Rahul", "Kiran"} is a subset of loan_customers
loan_customers = {"Rahul", "Kiran", "Asha", "Ramesh"}
print({'Rahul','Asha'}.issubset(loan_customers))
```

```python
#Check whether customers is a superset of {"Rahul", "Anita"}.
customers = {"Rahul", "Anita", "Kiran", "Sneha", "Vikram"}
print(customers.issuperset({'Rahul','Vikram'}))
```

```python
#Check whether loan_customers and new_cities are disjoint.
loan_customers = {"Rahul", "Kiran", "Asha", "Ramesh"}
new_cities = {"Mumbai", "Bangalore", "Delhi"}
print(loan_customers.isdisjoint(new_cities))
```
